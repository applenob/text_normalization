```python
das_pat = re.compile(r"^[a-zA-Z]+-[a-zA-Z]+$")

def has_dash(token):
    token = str(token)
    if das_pat.search(token):
        return True
    else:
        return False
```

```python
# strange_list = list("₀₁₂₃₄₅₆₇₈₉⁰¹²³⁴⁵⁶⁷⁸⁹፬")
strange_list = list("₀₁₂₃₄₅₆₇₈₉")
def is_strange(token):
    token = str(token)
    for c in strange_list:
        if c in token:
            return True
    return False
```

```python
train_df.loc[train_df["before"].apply(is_strange)]
```

```python
train_df.loc[np.logical_and(train_df["before"].apply(is_strange), train_df["class"] != "MEASURE")]
```

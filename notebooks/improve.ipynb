```python
def is_com(token):
    import re
    token = str(token)
    pt = re.compile(r"\.[a-zA-Z]{2,}")
    return True if pt.search(token) else False
```

```python
plain_df_f3 = plain_df_f2.loc[~plain_df_f2["before"].apply(is_com)]
plain_df_f3
```

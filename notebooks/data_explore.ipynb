## en_train.csv

```python
train_df = pd.read_csv('../input/en_train.csv')
train_df.head()
```

```python
# 查看所有类别class
train_df["class"].unique()
```

```python
labels = train_df["class"].unique()
```

```python
len(labels)
```

```python
for l in labels:
    print("label: ", l)
    train_df_l = train_df.loc[train_df["class"] == l]
    print("origin num: ", train_df_l.shape[0])
    equal_df = train_df_l.loc[train_df["before"] == train_df["after"]]
    print("equal num: ", equal_df.shape[0])
    print("rate: ", float(equal_df.shape[0]) / train_df_l.shape[0])
    print(train_df_l[:10])
```
